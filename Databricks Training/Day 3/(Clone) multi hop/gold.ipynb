```python
%sql
create or replace view acc_dev.alekhya_gold.top_customers as
(select customer_id, round(sum(total_amount)) as total_amount 
from acc_dev.alekhya_silver.sales_cleaned 
group by customer_id 
order by total_amount desc limit 3)
```
