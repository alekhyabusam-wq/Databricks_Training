```python
from pyspark.sql.functions import col
df.select(col("CustomerID").alias("cust_id"))
```

```python
df.select("CustomerID",col("TransactionID"),df.AccountType,df["AmountUSD"]).display()
```

```python
df.withColumnRenamed("CustomerID","cust_id").withColumnRenamed("TransactionID","transaction_id").display()
```

```python
df.withColumnsRenamed({"CustomerID": "cust_id", "TransactionID": "transaction_id"}).display()
```

```python
from pyspark.sql.functions import upper
df.withColumn("Branch", upper("Branch")).display()
```

```python
from pyspark.sql.functions import current_timestamp
df.withColumn("ingestion_date",current_timestamp()).display()
```
