###End to End ETL for financials

```python
df = spark.read.csv("/Volumes/acc_dev/naval/raw/financial_dataset.csv",header=True,inferSchema=True)
```

```python
df = df.withColumnsRenamed({"TransactionID": "transaction_id", "CustomerID": "customer_id","AccountType": "account_type","TransactionType": "transaction_type","AmountUSD": "amount_usd","Branch": "branch","TransactionDate": "transaction_date"})
```

```python
from pyspark.sql.functions import *
df = df.withColumn("ingestiondate", current_timestamp()).withColumn("path", col("_metadata.file_name"))
df.display()
```

```python
df.write.mode("overwrite").option("mergeSchema", "true").saveAsTable("acc_dev.alekhya.financial_dataset_py")
```
